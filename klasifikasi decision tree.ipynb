```python
df=pd.read_csv('predic_tabel.csv')
df.head()
```

```python
df.drop(['No', 'Jenis_Kelamin', 'Bekerja', 'Rumah_Tangga', 'Asuransi'], axis=1, inplace=True)
```

```python
df.info()
```

```python
labelencoder = LabelEncoder()
df['Usia'] = labelencoder.fit_transform(df['Usia'])
df['Merokok'] = labelencoder.fit_transform(df['Merokok'])
df['Aktivitas_Begadang'] = labelencoder.fit_transform(df['Aktivitas_Begadang'])
df['Aktivitas_Olahraga'] = labelencoder.fit_transform(df['Aktivitas_Olahraga'])
df['Penyakit_Bawaan'] = labelencoder.fit_transform(df['Penyakit_Bawaan'])
df['Hasil'] = labelencoder.fit_transform(df['Hasil'])
```

```python
df.head()
```

```python
X = df[['Usia', 'Merokok', 'Aktivitas_Begadang', 'Aktivitas_Olahraga', 'Penyakit_Bawaan']]
y = df['Hasil']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
sc = MinMaxScaler(feature_range=(0, 1))

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train, "\n")
print(X_test, "\n")
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
```

```python
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Classification Report : \n", classification_report(y_test, y_pred))
print("F1 Score : \n", f1_score(y_test, y_pred))
print("Precision Score : \n", precision_score(y_test, y_pred))
print("Recall Score : \n", recall_score(y_test, y_pred))
print("Accuracy Score : \n", accuracy_score(y_test, y_pred))
```
